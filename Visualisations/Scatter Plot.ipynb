```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Summative Data.csv")


def scatter_plot(inp_df, col1, col2):
    sns.jointplot(
        data=inp_df,
        x=col1,
        y=col2,
        kind="reg",
        joint_kws={"line_kws": {"color": "red"}},
    )


scatter_plot(df, "tenure", "Years_employed")
plt.show()
```
