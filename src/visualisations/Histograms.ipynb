```python
import pandas as pd
import matplotlib.pyplot as plt  # import library for 2D visualisations

df = pd.read_csv("StudentsPerformance.csv")

column_data = df["math score"]  # defines column for analysis

plt.figure(figsize=(10, 6))  # specifies size of the chart
plt.hist(
    column_data, bins=100, edgecolor="black"
)  # bin means how many bars in the chart
plt.title("Math Scores")
plt.xlabel("Score")
plt.ylabel("frequency")
plt.grid(True)  # adds a grid for improved visualisation
plt.show()
```
