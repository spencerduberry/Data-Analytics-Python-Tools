```python
import pandas as pd
import matplotlib.pyplot as plt #import library for 2D visualisations

df = pd.read_csv('Summative Data.csv')

column_data = df['balance'] #defines column for analysis

plt.figure(figsize=(10, 6))  #specifies size of the chart
plt.hist(column_data, bins=10, edgecolor='black')  #bin means how many bars in the chart
plt.title('Balance distribution')
plt.xlabel('balance')
plt.ylabel('frequency')
plt.grid(True)  #adds a grid for improved visualisation
plt.show()
```
